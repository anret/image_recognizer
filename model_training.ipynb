```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_digits
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = load_digits()
print(dataset.keys())
print(dataset)
```

```python
index = [f's{i}' for i in range(1,len(dataset['data'])+1)]
columns = dataset['feature_names']
X = pd.DataFrame(data=dataset['data'], index=index, columns=columns)
y = pd.DataFrame(data=dataset['target'])
X
```

```python
sample = 2
plt.imshow(X.iloc[sample].to_numpy().reshape(8,8))
y.iloc[sample]
```

```python
model = RandomForestClassifier()
model.fit(X.to_numpy(), y.to_numpy())
```

```python
model.score(X,y)
```

```python
sample = X.iloc[0].to_numpy().reshape(1,-1)
print(sample)
pred = model.predict(sample)
pred
```

```python
import pickle
pickle.dump(model, open('models/forest', 'wb'))
```
